```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Creating a blank image (Black background)
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

### Explanation: 
* np.zeros creates an array filled with zeros (black image)
* Shape (512, 512, 3) means a 512x512 image with 3 color channels (RGB)
* dtype=np.int16 defines the type of values in the array (important for image processing)


```python
# Displaying the blank image
plt.imshow(blank_img)
plt.title("Blank Image")
plt.show()
```

```python
# Drawing a rectangle
cv2.rectangle(blank_img, pt1=(384, 10), pt2=(500, 150), color=(0, 255, 0), thickness=10)
```

### Explanation:
* pt1: Top-left corner of the rectangle
* pt2: Bottom-right corner
* color: Green (0,255,0 in BGR format)
* thickness: 10 pixels


```python
# Displaying the updated image
plt.imshow(blank_img)
plt.title("Rectangle Drawn")
plt.show()
```

```python
# Drawing a circle
cv2.circle(img=blank_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness=8)
cv2.circle(img=blank_img, center=(400, 400), radius=50, color=(255, 0, 0), thickness=-1)  # Filled circle
```

### Explanation:
* center: (x, y) coordinate of the circle center
* radius: Circle radius in pixels
* color: Blue (255,0,0 in BGR format)
* thickness=-1 means fill the circle with color


```python
# Displaying the updated image
plt.imshow(blank_img)
plt.title("Circles Drawn")
plt.show()
```

```python
# Drawing a line
cv2.line(blank_img, pt1=(0, 0), pt2=(512, 512), color=(102, 255, 255), thickness=5)

```

### Explanation:
* pt1: Start point
* pt2: End point
* color: Cyan (102,255,255 in BGR format)
* thickness: 5 pixels


```python
# Displaying the updated image
plt.imshow(blank_img)
plt.title("Line Drawn")
plt.show()
```

```python
# Adding text
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Hello', org=(10, 500), fontFace=font, fontScale=4, 
            color=(255, 255, 255), thickness=3, lineType=cv2.LINE_AA)
```

### Explanation:
* text: String to display
* org: Bottom-left corner where the text starts
* fontFace: Font type (cv2.FONT_HERSHEY_SIMPLEX)
* fontScale: Size of text
* color: White (255,255,255)
* thickness: Thickness of the text
* lineType: cv2.LINE_AA for anti-aliased text (smooth edges)


```python
# Displaying the updated image
plt.imshow(blank_img)
plt.title("Text Added")
plt.show()
```

```python
# Creating another blank image for polygon drawing
polygon_img = np.zeros(shape=(512, 512, 3), dtype=np.int32)

```

```python
# Defining polygon vertices
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)

```

```python
# Reshaping the vertices for OpenCV format
pts = vertices.reshape((-1, 1, 2))  # Reshape to (N,1,2) for polylines

```

### Explanation:
* reshape(-1,1,2) converts 2D array into a shape OpenCV expects for polygons
* -1 means "infer number of rows automatically"
* 1 keeps individual points in their own group
* 2 represents x and y coordinates



```python
# Drawing the polygon
cv2.polylines(polygon_img, [pts], isClosed=True, color=(0, 255, 255), thickness=5)

```

### Explanation:
* isClosed=True means it will close the shape automatically
* color: Yellow (0,255,255)
* thickness: 5 pixels


```python
# Displaying the polygon
plt.imshow(polygon_img)
plt.title("Polygon Drawn")
plt.show()
```
