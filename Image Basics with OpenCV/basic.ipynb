Introduction
OpenCV (Open Source Computer Vision) is a powerful library mainly used or real-time computer vision applications.It was created by Intel in 1999 and is written in C++, but is widely used with Python bindings.

## Why Use OpenCV
* **Performance:** Written in C++ for high-speed image processing
* **Rich Features:** Includes built-in algorithms for object detection, tracking and transformation.
* **EasyIntegration:** Works seamlessly with Numpy and Matplotlib for visualization.

```python
##Import  necessary libararies
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
# Ensure inline plotting in Jupyter Notebook
%matplotlib inline
```

```python
## 1.Loading image
#read image using OpenCV
image= cv2.imread('../Numpy and Image Basics\s.jpg')
```

```python
# Check if the image is loaded correctly
if image is None:
    raise FileNotFoundError("Error: Image not found. Check the file path.")
```

```python
## Display the image
plt.imshow(image) #This will show different color because OpenCV use BGR format
plt.title("BGR Image")
plt.show()
```

```python
##2. Fixing Color Format
#OpenCV loads image in BGR format; Matplotlib excepts RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Corrected RGB Image")
plt.show()

```

```python
## 3. Converting to Grayscale
# Convert image to grayscale
image_gray = cv2.imread('../Numpy and Image Basics\s.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(image_gray, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```

```python
## 4.Checking the image properties
print("Original Image Shape:", image.shape)  # (Height, Width, Channels)
print("Grayscale Image Shape:", image_gray.shape)  # (Height, Width)
```

```python
## 5.Resizing an Image
#Resize image to 100x100 pixels

image_resized = cv2.resize(image_rgb, (100,100))

plt.imshow(image_resized)
plt.title("Resized Image (100x100)")
plt.show()
```

```python
# Resize while maintaining aspect ratio
scale_percent = 50  # Scale down by 50%
width = int(image_rgb.shape[1] * scale_percent / 100)
height = int(image_rgb.shape[0] * scale_percent / 100)
image_resized_aspect = cv2.resize(image_rgb, (width, height))

plt.imshow(image_resized_aspect)
plt.title("Aspect Ratio Preserved Resize")
plt.show()
```

```python
## 6. Flipping an IMage
#Flip vertically(0), horizontally(1), both(-1)

image_flipped = cv2.flip(image_rgb, 1)
plt.imshow(image_flipped)
plt.title("Ia=mage Flipped Horizonatlly")
```

```python
image_flipped= cv2.flip(image_rgb,  0)
plt.imshow(image_flipped)
plt.title("Image Flipped Vertically")
plt.show()
```

```python
image_flipped = cv2.flip(image_rgb, -1)
plt.imshow(image_flipped)
plt.title("Image Flipped both Vertically and Horizontally")
plt.show()
```

```python
## 7. Saving an Image
cv2.imwrite('modified_image.jpg', image)
print("Image saved successfully as modified_image.jpg")
```

## Conclusion
- OpenCV provides robust functions for image handling.
- We must adjust colors (BGR  RGB) for proper visualization in Matplotlib.
- Image resizing, flipping, and drawing operations are simple yet powerful.
- Saving and loading images efficiently is essential in computer vision tasks.
