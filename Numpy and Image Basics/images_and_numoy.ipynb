## Introduction
This notebook covers fundamental concepts of handling images using python. We will use libraries such as Numpy, Pillow(PIL), and matplotlib to read, process, and visualize images.

## Why These Librarries?
### Pillow (PIL) - Python Image Library
* Numpy cannot directly open images files like .png or .jpg
* Pillow is used to **load, manipulate, save images** in Python
* Provides easy integration with Numpy for numerical processing.


### Matplotlib
* used ofr **visualizing images**.
* Allows us to display different color channels seperately.
* Provides colormap functionalities for graysscale representation.

### Numpy
* Converts images into arrays for **efficient computation**.
* Helps in extracting and modifying pixel values.

```python
### importing required libraries
import numpy as np
import matplotlib.pyplot as plt 
from PIL import Image
```

```python
#Enable inline plotting
%matplotlib inline
```

```python
## Loading and Displaying an images
# Open an image uisng Pillow
pic = Image.open("Lenna.png")
```

```python
## Display the image
plt.imshow(pic)
plt.axis("off") # Hide axes
plt.show()
```

```python
### Understanding Image Types
#Check the type of image object
print(type(pic)) 
```

```python
#Covert image into NumPy array
pic_arr = np.array(pic)
```

```python
# Check the shape of the array
print(pic_arr.shape) # Output: (height, width, 3)
```

```python
### Extracting Color channels
## Ecah pixel in an image is represented by three values (R,B,G). We can extract each channel seperately
#Create copy of the image array
pic_cpy = pic_arr.copy()
```

```python
# Display only the red channel
plt.imshow(pic_cpy[:,:,0], cmap= 'gray') # Converting Red channel to grayscale for better visualization
plt.title("Red Channel")
plt.axis("off")
plt.show()
```

### Why Extract Individual Channels?
* Useful for **color-based object detection**.
* Helps in **color blindness adaptation**.
* Enhances **edge detection** by analysing single channels.

```python
## Making a Channel Zero (Removing a Color)
## We can remove a particular channel by setting all its value to zero.

pic_no_red = pic_arr.copy()
pic_no_red[:,:,0]= 0 #set red channel to 0
```

```python
plt.imshow(pic_no_red)
plt.title("Image without Red Channel")
plt.axis("off")
plt.show()
```

```python
## Making a Channel Zero (Removing a Color)
## We can remove a particular channel by setting all its value to zero.

pic_no_green = pic_arr.copy()
pic_no_green[:,:,1]= 0 #set green channel to 0
```

```python
plt.imshow(pic_no_green)
plt.title("Image without Green channel")
plt.axis("off")
plt.show()
```

```python
## Making a Channel Zero (Removing a Color)
## We can remove a particular channel by setting all its value to zero.

pic_no_blue = pic_arr.copy()
pic_no_blue[:,:,2]= 0 #set blue channel to 0
```

```python
plt.imshow(pic_no_blue)
plt.title("Image without Blue channel")
plt.axis("off")
plt.show()
```

## Why Normalize Images?
- Pixel values range from **0 to 255**.
- Normalization brings values between **0 and 1**.
- Helps machine learning models train efficiently.


```python
pic_normalized = pic_arr / 255.0
print("Normalized Pixel Values:", pic_normalized[:2, :2, :])  # Display first few pixel values
```
